# E-Commerce Negotiation Platform

## One-Sentence Summary
This app offers a platform for a customer to negotiate with retailer company's computer agent when purchase online

## The Problem solved
The app can help a company to sell products with a flaxible price corresponding to customer's personal requirement

## User Types
E-Commerce customers

## Core Features

- Administrator can setup what product to negotiate
- Set up user and agent's personal preference about products
- Exchange offers for product price or other features
- Accept offers if user liked it or satisfied pre-defined agent's preference

## What Users See

### Screen1: userDemographic
Ask user demographic information and save
### Screen2.1: product and features
a popup window ask administrator what product to negotiate and what features can be under negotiation (such as laptop, features are CPU, hard drive, RAM, screen size)
### Screen2.2: options
a popup window ask administrator what options does each feature have. (such as for feature RAM, options are 1G, 5G, 10G, 50G)
### Screen2.3: agent name and pic
a popup window ask administrator to upload profile image and make up a name for agent
### Screen3.1: preference
Set up weight distribution for all the features. All the feature's weight should add up to 100% (for example, 50% weight assigned to price, 30% to CPU, 10% to hard drive, 8% to screen size, 2% to RAM.)
### Screen3.2: featureOptionValues
Set up utility values for the options of each feature (for example, for RAM, the seller computer agent's utility is 100 for 1G, 80 for 5G, 70 for 10G, 60 for 50G. The highest value should be 100)
### Screen3.3: concessionCurve
Setup concession curve for agents. the curve could be convex, concase, or linear. the curve depends on admin inputting a formula or uploading a picture. The page should also ask admin to input a formula (using all the features as variables) to calculate the utilty value for all the offers.
### Screen4: offerCompose
In the offer-composing page, the user should choose one option for each feature and then give a price using a slide bar. the page should show the utility value calculation reuslt. The utility is calculated depending on the features' weight (screen3.1), the utility values for the chosen options (screen3.2), and the formula input by the admin on screen 3.3.
### Screen5: negotiation
The user and agent exchange offers around the features.
on this page, each offer is shown as one row. each column is one feature. column content should show the chosen options. the first column should be the agent's or the user's profile image. the last two columns should be the price and utility value of each offer.
At the botton of this page, there shuold be 3 buttons: "accept the offer", "compose a counter offer", and "end the negotiation".

## Example Scenario

Lisa is the admin of the experiment. When she click "admin log in" and login with her email. she can access admin page. she can click "start new experiment" button, a page opens and ask her what topic she wants to use for the experiment and what features should include for this topic in the negotiation experiment (screen2.1. for example, the topic could be the purchase of a laptop. then features could be: price, RAM, hard drive, CPU, screen size). Then, this page should ask her, for every single feature, what options do users have (screen2.2. for example, for RAM, the options could be 1G, 5G, 10G, and 50G). Then this page (screen2.3) should ask her to upload a profile image and make up a name for this agent.
As the admin, Lisa should set up the preference/weight of the features for such topic (screen3.1  for example, 50% weight assigned to price, 30% to CPU, 10% to hard drive, 8% to screen size, 2% to RAM. all features should add up to be 100%). For each feature, she also need to set up the utility values for each option (Screen3.2  for example, for RAM, the seller computer agent's utility is 100 for 1G, 80 for 5G, 70 for 10G, 60 for 50G. The highest value should be 100. For price, the page should use a range instead of options).
The admin should also set up the concession curve and input a formula to calculate utility for agent (screen3.3).  
Then, the page (screen3.3 ) should also ask if she wants to create multiple agents. if true, the page should prompt she go though the whole process to create another agent.

For user of this website, A user first click "start" button to start a negotiation. The demographic page (screen1) ask for user's name, age, email, gender, etc.. Then the screen3.1 popup ask for user's preference/weight. Screen3.2 ask for user's options utility values. Then the user can start a negotiation with computer agent.

In the negotiation, there should be offer exchange. the agent always give the first offer. The user should see screen5. they could choose accept the agent's last offer and end the negotiation, compose a counter offer, or end the negotiation. If the user choose to compose a counter offer, the screen4 should show to let user compose an offer.
In the offer-composing page (screen 4), the user should choose one option for each feature and then give a price using a slide bar. the page should show the utility value calculation reuslt. The utility is calculated depending on the features' weight (screen3.1), the utility values for the chosen options (screen3.2), and the formula input by the admin on screen 3.3.
After the user click "send offer" button on offer composing page, the user should see the negotiation page (screen5).
After the negotiation is ended, a thank you page should show.
